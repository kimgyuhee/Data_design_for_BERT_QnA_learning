```python
import csv
import matplotlib.pyplot as plt

class CsvData() :
    def open_csv(self) :
        file = open('subwaytime.csv')
        data = csv.reader(file)

        time = next(data)
        on_off = next(data)

        s_in = [0] * 24 # 승차인원 합계
        s_out = [0] * 24 # 하차인원 합계

        mx = [0] * 24 # 시간대별 가장 많은 승차 값 저장할 리스트
        mx_station = [''] * 24 # 시간대별 가장 많은 승차 역 저장할 리스트


        mn = [100] * 24 # 시간대별 가장 많은 하차 값 저장할 리스트
        mn_station = [''] * 24 # 시간대별 가장 많은 하차 역 저장할 리스트

        for row in data :
            row[4:] = map(int, row[4:])
            for i in range(24) :
                on_result = row[i*2+4]
                off_result = row[i*2+5]
                s_in[i] +=row[2*i+4]
                s_out[i] +=row[2*i+5]
                if on_result > mx[i] :
                    mx[i] = on_result
                    mx_station[i] = row[3] + '(' + row[1] + ')' + str((i+4)%24) + '시'
                if off_result > mn[i] :
                    mn[i] = off_result
                    mn_station[i] = row[3] + '(' + row[1] + ')' + str((i+4)%24) + '시'

        time_list = []
        for i in range(24) :
            time_list.append((i+4)%24)
        file.close()

        return s_in, s_out, mx,mx_station, mn, mn_station, time_list

class PlotVisual() :
    def on_off_visual_plot(self, s_in, s_out, time_list) : # 꺽은선 그래프
        plt.rc('font', family='Malgun Gothic')
        plt.title('지하철 시간대별 승하차 인원 추이')
        plt.plot(s_in, label= '승차')
        plt.plot(s_out, label= '하차')
        plt.legend()
        plt.xticks(range(24), time_list)
        plt.show()
    
    def on_visual_plot(self, mx, mx_station) : # 막대 그래프
        plt.rc('font', family='Malgun Gothic')
        plt.title('시간대별 승차 인원이 가장 많은 인원 추이')
        plt.plot(mx, label='승차', color = 'orange')
        plt.legend()
        plt.xticks(range(24), mx_station, rotation=90)
        plt.show()
    
    def off_visual_plot(self, mn, mn_station) : # 막대 그래프
        plt.rc('font', family='Malgun Gothic')
        plt.title('시간대별 하차 인원이 가장 많은 인원 추이')
        plt.plot(mn, label='하차')
        plt.legend()
        plt.xticks(range(24), mn_station, rotation=90)
        plt.show()
        
        
class BarVisual() :
    def on_visual_bar(self, mx, mx_station) : # 막대 그래프
        plt.rc('font', family='Malgun Gothic')
        plt.title('시간대별 승차 인원이 가장 많은 인원 추이')
        plt.bar(range(24), mx, label='승차', color = 'orange')
        plt.legend()
        plt.xticks(range(24), mx_station, rotation=90)
        plt.show()
    
    def off_visual_bar(self, mn, mn_station) : # 막대 그래프
        plt.rc('font', family='Malgun Gothic')
        plt.title('시간대별 하차 인원이 가장 많은 인원 추이')
        plt.bar(range(24), mn, label='하차')
        plt.xticks(range(24), mn_station, rotation=90)
        plt.legend()
        plt.show()
        
    def on_off_visual_bar(self, s_in, s_out, time_list) :
        def minus(n) :
            return -n
        m_s_out = list(map(minus, s_out))
        plt.rc('font', family='Malgun Gothic')
        plt.rcParams['axes.unicode_minus'] = False
        plt.title('시간대별 승하차 인원 추이')
        plt.bar(range(24), s_in, label= '승차')
        plt.bar(range(24), m_s_out, label= '하차')
        plt.legend()
        plt.xticks(range(24), time_list)
        plt.show()
        
        
class TextCheck() :
    def on_text_chcek(self, time_list, mx, mx_station) :    
        print('========== 2019년 1월 시간별 승차 통계 ==========')
        for i in range(24) :
            print('{}시에 승차가 가장 많은 역은 {}이고 그 값은 {:,}명 입니다.'.format( time_list[i],  mx_station[i], mx[i]))

    def off_text_chcek(self, time_list, mn, mn_station) :    
        print('========== 2019년 1월 시간별 하차 통계 ==========')
        for i in range(24) :
            print('{}시에 하차가 가장 많은 역은 {}이고 그 값은 {:,}명 입니다.'.format( time_list[i],  mn_station[i], mn[i]))

    def on_off_text_check(self, time_list, s_in, s_out) :
        print('========== 2019년 1월 시간별 승하차 합계 통계 ==========')
        for i in range(24) :
            print('{}시에 승차 인원의 합은 {:,}명이고, 하차 인원의 합은 {:,}명 입니다.'.format(time_list[i], s_in[i], s_out[i]))
            
            
def choice_option() :
    option_text = '''
원하는 메뉴를 선택해주세요.
1. 막대 그래프로 보기
2. 꺽은선 그래프로 보기
3. 텍스트 형식으로 보기
번호 선택 -> '''
    
    while True :
        option = input(option_text)
        if not option.isdecimal() :
            print('1~3번 사이의 숫자를 입력해주세요.\n')
            continue
        else :
            option = int(option)
            if option == 1 :
                return 1
            elif option == 2 :
                return 2
            elif option == 3 :
                return 3
            else :
                print('1~3번 사이의 숫자를 입력해주세요.\n')
                continue

while True :
    answer = input('''
===== 원하는 프로그램을 입력해주세요 =====
1. 시간대별 승차 통계
2. 시간대별 하차 통계
3. 시간대별 승하차 인원 추이
4. 프로그램 종료
번호 선택 -> ''')
    
    if not answer.isdecimal() :
        print('1~4번 사이의 숫자를 입력해주세요.\n')
        continue
    else :
        answer = int(answer)
        if answer == 4 :
            print('\n\n========== 프로그램을 종료합니다. ==========\n')
            break
            
        csvdata = CsvData()
        # csv파일에서 원하는 데이터 추출하기
        s_in, s_out, mx,mx_station, mn, mn_station, time_list = csvdata.open_csv()
        barv = BarVisual()
        text_check = TextCheck()
        plotv = PlotVisual()
        num = choice_option()
        if answer == 1 :
            if num == 1 :
                barv.on_visual_bar(mx, mx_station)
            elif num == 2 :
                plotv.on_visual_plot(mx, mx_station)
            else :
                text_check.on_text_chcek(time_list, mx, mx_station)      
        
        elif answer == 2 :
            if num == 1 :
                barv.off_visual_bar(mn, mn_station)
            elif num == 2 :
                plotv.off_visual_plot(mn, mn_station)
            else :
                text_check.off_text_chcek(time_list, mn, mn_station)
                    
        elif answer == 3 :
            if num == 1 :
                barv.on_off_visual_bar(s_in, s_out, time_list)
            elif num == 2 :
                plotv.on_off_visual_plot(s_in, s_out, time_list)
            else :
                text_check.on_off_text_check(time_list, s_in, s_out)
        
        else :
            print('1~4번 사이의 숫자를 입력해주세요.\n')
            continue
                
```
