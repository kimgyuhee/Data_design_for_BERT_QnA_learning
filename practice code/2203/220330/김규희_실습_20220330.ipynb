```python
import csv
import time


print('=='*40)
print('시간별 승차인원이 가장 많거나 적은 역을 찾는 프로그램입니다.')
print('=='*40)
time.sleep(1)

class Station() :
    
     # 시간대를 입력하는 함수
    def input_time(self, what) :
        while True :
            time = input('몇시에 승차 인원이 가장 {} 역을 찾으시나요? '.format(what))
            if not time.isdecimal() : # 입력 받은 값이 숫자가 아닐 때
                print('0~23 사이의 숫자를 입력해주세요.\n')
                continue
            else : # 입력 받은 값이 숫자 일 때
                time = int(time) 
                if time > 23 or time < 0 :
                    print('0~23 사이의 숫자를 입력해주세요.\n')
                    continue
                else :
                    break

        return time   
    
    # 입력 받은 시간대의 인덱스를 찾는 함수
    def find_index(self, t) :
        file = open('subwaytime.csv')
        time_find_data = csv.reader(file)
        data = csv.reader(file)
        # num = 50
        subway_time = next(data)
        on_off = next(data)

        find_time = []
        global num
        for i in range(len(subway_time)) :
            value = subway_time[i].split(':')[0]
            if value.isdecimal() :
                value = int(value)
            find_time.append(value)
        
        for i in range(len(find_time)) :
            if t == find_time[i] :
                num = i
        
        return num

    def continue_system(self) :
        while True :
            answer = input('프로그램을 계속 실행하시겠습니까? [Y, N] ')
            if answer.upper() == 'Y' :
                return True
            elif answer.upper() == 'N' :
                return False
            else :
                print('Y와 N 중에 입력해주세요.\n')
                continue
    
    
class HourlyMaxStation(Station) :

    def __init__(self) :
        what = '많은'
        t = super().input_time(what)
        num = super().find_index(t)
        
    # csv 파일을 열어 원하는 데이터를 추출하기 위한 함수
    def open_csv(self, num) :
        file = open('subwaytime.csv')
        data = csv.reader(file)

        time = next(data)
        on_off = next(data)

        mx = 0
        mx_station = ''
        
        for row in data :
            row[4:] = map(int, row[4:])
            value = row[num]
            if value > mx :
                mx = value
                mx_station = row[3]+'('+row[1]+')'

                
        file.close()
        return mx, mx_station

    def __str__(self) :
        mx, mx_station = self.open_csv(num)
        text = '\n{}시에 승차가 {} 역은 {} 입니다\n'.format(t, '많은' ,mx_station)
        text = text + '승차 인원은 {:,}명 입니다.\n'.format(mx)
        return text
    
class HourlyMinStation(Station) : 

    def __init__(self) :
        what = '적은'
        t = super().input_time(what)
        num = super().find_index(t)
        
    # csv 파일을 열어 원하는 데이터를 추출하기 위한 함수
    def open_csv(self, num) :
        file = open('subwaytime.csv')
        data = csv.reader(file)

        time = next(data)
        on_off = next(data)

        mn = 10000000
        mn_station = ''
        
        for row in data :
            row[4:] = map(int, row[4:])
            value = row[num]
            if value < mn :
                mn = value
                mn_station = row[3]+'('+row[1]+')'

                
        file.close()
        return mn, mn_station
    
    def __str__(self) :
        mn, mn_station = self.open_csv(num)
        text = '\n{}시에 승차가 {} 역은 {} 입니다\n'.format(t, '적은' ,mn_station)
        text = text + '승차 인원은 {:,}명 입니다.\n'.format(mn)
        return text
    
while True :
    
    mx_mn = input('1. 최대\t\t2. 최소 -> ')
    if not mx_mn.isdecimal() :
        print('1과 2중 선택해 숫자를 입력해주세요.\n')
    else :
        mx_mn = int(mx_mn)
        if mx_mn == 1 :
            station = HourlyMaxStation()
            print(station)
            
            if station.continue_system() :
                print()
                continue
            else :
                print('\n프로그램을 종료합니다.')
                break
                
        elif mx_mn == 2 :
            station = HourlyMinStation()
            print(station)
            
            if station.continue_system() :
                print()
                continue
            else :
                print('\n프로그램을 종료합니다.')
                break
        else :
            print('1과 2중 선택해 숫자를 입력해주세요.\n')
```
