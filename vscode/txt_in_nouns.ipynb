```python
# 기존 txt 파일에서 명사만 추출

from konlpy.tag import Okt
import sys 

class NounFilter:
    def __init__(self): # NounFilter 프로그램 시작시 호출되는 welcoming 함수 
        print('='*60)
        print('\t\t  명사 추출 프로그램을 시작합니다.')
        print('='*60)
        
    def file_dir(self): # 명사 추출 이후에 txt 파일을 저장시킬 경로를 지정받는 함수 
        print('txt 파일을 저장합니다. 저장 경로를 지정해주세요.\n예) D:\\2th_project\\W_test_country\\ ')
        self.file_directory = input('저장 파일 경로: ') # 파일 경로는 file_directory 변수에 담습니다.
        print('='*60)
        return self.file_directory
    
    def file_n(self):  # 저장시킬 txt 파일의 이름을 지정받는 함수 
        print('txt 파일의 이름을 지정해주세요.\n예) test')
        self.file_name = input('txt 파일명 지정: ') # 입력 받은 파일명은 file_name 변수에 담습니다. 
        print('='*60)
        return self.file_name
                                                                # 앞서 입력받은 파일 경로와 파일명을 종합하고 사용자에게 안내해주는 함수
    def confirming_file_name(self, file_directory, file_name):  # file_directory: 사용자 지정의 파일 경로, file_name: 사용자 지정의 파일 이름
        print('txt 파일을 아래와 같이 저장합니다.')
        print(f'경로: {file_directory}\n이름: {file_name}')
        print('='*60)
        self.txt_file = file_directory + file_name + '.txt'
        return self.txt_file                                    # 파일 경로와 파일명 그리고 확장명이 더해진것을 txt_file 변수에 담습니다.
    
    def open_raw_data(self): # 명사만 추출할 txt 파일을 불러오는 함수 
        print('명사를 추출할 데이터 파일의 경로를 입력해주세요.\n예) D:\\2th_project\\country\\F_Russia.txt ')
        self.raw_data = input('데이터 파일 경로: ') # txt 파일의 경로를 raw_data 변수에 담습니다. 
        print('='*60)
        return self.raw_data 
                                      # 명사만 추출할 txt 파일을 읽어주는 함수 
    def read_file(self, raw_data):    # raw_data: txt 파일의 경로를 담고있습니다. 
        with open(self.raw_data, 'r', encoding='utf-8') as self.f:
            self.text = self.f.read() # 읽어온 txt 파일은 text 변수에 담습니다. 
        return self.text
                                           # 새로운 txt 파일에 형태소 분석 후 명사만 담아줄 함수
    def write_file(self, text, txt_file):  # text: 말뭉치가 담긴 text 파일(읽기모드), txt_file: 파일경로와 파일명을 담고있는 함수
        self.okt = Okt()                   # 형태소 분석기인 Okt를 okt 변수에 담습니다. 
        self.noun = self.okt.nouns(text)   # text 변수로 불러온 데이터에서 명사만 추출합니다. text: 말뭉치가 담긴 text 파일(읽기모드)
        
        self.orig_stdout = sys.stdout
        with open(txt_file, 'w', encoding='utf-8') as self.file:
            sys.stdout = self.file
            
            for i in self.noun:
                if len(i) > 1 and i != '문장': # 글의 길이가 1자 이하이거나 txt 문서에 '문장'으로 구분지어둔 단어는 생략합니다. 
                    print(i)

            sys.stdout = self.orig_stdout
            
        print('명사 추출이 완료되었으며 파일이 정상적으로 저장되었습니다.')

noun_f = NounFilter()              # NounFilter 클래스의 인스턴스
file_directory = noun_f.file_dir() # 파일 경로를 지정받는 함수
file_name = noun_f.file_n()        # 파일 이름을 지정받는 함수 
txt_file = noun_f.confirming_file_name(file_directory, file_name) # 저장될 파일 경로와 이름을 보여주는 함수 file_directory: 파일 경로, file_name: 파일명
raw_data = noun_f.open_raw_data()  # txt 파일을 열어줄 함수
text = noun_f.read_file(raw_data)  # txt 파일을 읽어줄 함수 raw_data: txt 파일의 경로를 담고있는 변수 
noun_f.write_file(text, txt_file)  # 명사만 추출하여 txt 파일을 써주는 함수 text: 말뭉치가 담긴 text 파일(읽기모드), txt_file: 파일경로와 파일명을 담고있는 함수
    

```
