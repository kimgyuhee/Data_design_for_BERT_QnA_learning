```python
import pandas as pd
import os
from tqdm import tqdm

# xlsx 파일들이 존재하는 폴더 불러오는 함수
def input_xlsx() :
    while True :
        path = input('xlsx 파일들이 존재하는 폴더를 입력해주세요.')
        if not os.path.isdir(path) :
            print('해당 파일이 존재하지 않습니다. 다시 입력해주세요.\n')
            continue
        else :
            # print('해당 파일이 존재하여 {} 폴더를 불러옵니다.\n'.format(path))
            break
    return path
        
# xlsx 리스트 목록을 보여주기 위한 함수
def output_xlsx_list(dir_path) :
    # 자신의 컴퓨터에 있는 xlsx 파일 폴더 넣어주기
    # dir_path = 'D:\\2th_project\\xlsx'

    xlsx_list = []
    files = os.listdir(dir_path)

    for i in files :
        if '.xlsx' or 'xls' in i :
            value = i.split('.')[0]
            xlsx_list.append(value)

    text = '--'*44
    for i in range(len(xlsx_list)) :
        if i%4 == 0 :
            text = text +'\n'
        text = text + str(i) + '. '+xlsx_list[i]+' \t\t'
        
    text = text+'\n'+'--'*44+'\n'
    return text, xlsx_list
     

def choice_country(text, xlsx_list) :
    while True :
        choice = input(text+'\n문장을 합치기 원하는 나라의 숫자를 입력해주세요 -> ')
        if not choice.isdecimal() : # 숫자를 입력하지 않았을 경우 - 예외처리
            print('0~{}사이의 숫자를 입력해주세요.\n'.format(len(xlsx_list)))
            continue
        else : # 숫자를 입력했을 경우
            choice = int(choice)
            if choice < 0 or choice > len(xlsx_list)-1 :
                print('0~{}사이의 숫자를 입력해주세요.\n'.format(len(xlsx_list)))
                continue
            else :
                break
                
    return choice, xlsx_list[choice]
            
    
def open_xlsx(country, path) :
    df = pd.read_excel(path+'\\'+country+'.xlsx')
    sent_count = len(df.index)
    sentence = []
    for i in range(sent_count) :
        value = df.iloc[[i], [1]]
        #print(value)
        # print(type(value))
    w = df['sentence']
    mystr = ''
    for i in tqdm(range(len(w))):
        if i == 0 :
            txt_name = w[i]
        else :
            mystr += str(w[i]) + '\n'

    print(mystr)
    
    return mystr, txt_name

def write_txt(text, txt_name) :
    dir_path = 'D:\\2th_project\\check\\'
    if not os.path.isdir(dir_path) :
        os.makedirs(dir_path)
    file_path = dir_path+txt_name+'.txt'
    file = open(file_path, 'w', encoding='utf-8');
    file.writelines(text)
    print()
    print('{} 문장들이 합쳐진 내용이 {} 파일에 저장되었습니다.\n'.format(txt_name, file_path))
    file.close()
    
    
    
# 함수 호출
path = input_xlsx()
text, xlsx_list = output_xlsx_list(path)
if len(xlsx_list) == 0 :
    print('{} 폴더에 xlsx 파일이 존재하지 않습니다.'.format(path))
else :
    num, country = choice_country(text, xlsx_list)
    text, txt_name = open_xlsx(country, path)
    write_txt(text, txt_name)
```
