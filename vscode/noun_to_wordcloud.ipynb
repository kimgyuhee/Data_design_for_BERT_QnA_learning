```python
# 마스킹 이미지로 워드클라우드

from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import sys 
from PIL import Image


def open_txt_file():     # 명사가 담긴 txt 파일을 읽어오는 함수
    txt_file = input('불러올 txt 파일의 위치를 입력하세요:\n기본 경로) /Users/deankim/Desktop/Developer/Project/noun_text/W_Russia.txt')
    if txt_file == '':   # 파일의 경로가 입력되지 않는 경우 기본 경로를 대신 입력합니다. 
        print('txt 파일 경로를 입력하지 않아 기본 경로에서 파일을 사용합니다.')
        txt_file = '/Users/deankim/Desktop/Developer/Project/noun_text/W_Russia.txt'
    txt = open(txt_file, 'r', encoding='utf-8') 
    lines = txt.read()   # lines 변수를 호출하면 txt 파일을 전체 읽습니다. 
    words = str(lines)   # 워드클라우드에서 읽을수있도록 lines 변수를 string으로 변환해줍니다. 
    txt.close()
    
    return words         # txt 파일의 내용을 담고있는 words 변수를 내보냅니다.

def open_img_file():     # 마스킹 이미지로 사용될 img 파일의 경로를 입력받는 함수
    img_file = input('마스킹 이미지로 사용할 파일의 경로를 입력하세요:\n기본 경로) /Users/deankim/Desktop/bear.jpeg')
    if img_file == '':   # 경로 입력 누락시 기본 경로를 입력 시킵니다.
        print('마스킹 이미지 파일의 경로를 입력하지 않아 기본 경로에서 파일을 사용합니다.')
        img_file = '/Users/deankim/Desktop/bear.jpeg'
    desg_mask = np.array(Image.open(img_file)) # 사용자가 입력한 img 파일을 desg_mask 변수에 담습니다. 
    
    return desg_mask

def set_font_path():     # 폰트 경로를 입력 받는 함수
    path = input('워드클라우드에서 사용할 폰트의 경로를 입력하세요:\n기본 경로) /Users/deankim/Desktop/Developer/Project/haenam.otf')
    if path == '':       # 폰트 경로가 입력되지 않은 경우 기본 경로를 입력 시킵니다.
        print('폰트 경로를 입력하지 않아 기본 경로에서 파일을 사용합니다.')
        path = '/Users/deankim/Desktop/Developer/Project/haenam.otf'
        return path     # 폰트 경로가 담긴 path 변수를 내보냅니다. 
    
    return path
                                         # txt 문서를 워드클라우드화 해주는 함수
def wordclouder(path, desg_mask, words): # path: 폰트 경로, desg_mask: 마스킹 이미지 경로, words: txt 내용(명사)
    wordcloud = WordCloud(font_path=path, background_color="white", max_font_size=100, mask=desg_mask).generate(words)
    print('성공적으로 워드클라우드 변환을 완료하였습니다.')
    
    return wordcloud                     # words에 담고있던 명사들을 워드클라우드화하고 이를 wordcloud 변수에 담아 내보냅니다.
        
def file_path_for_png():  # 워드클라우드화 이후 이미지 파일의 저장 경로를 지정 받는 함수 
    png_path = input('png 파일로 저장할 파일 경로를 입력하세요:\n기본 경로) /Users/deankim/Desktop/Developer/Project/test.png')
    if png_path == '':    # 이미지 파일 저장 경로가 누락시 기본 경로를 입력시킵니다.
        print('png 파일 경로를 입력하지 않아 기본 경로에서 파일을 사용합니다.')
        png_path = '/Users/deankim/Desktop/Developer/Project/test.png'
        return png_path   # 이미지 파일 저장 경로를 png_path 변수에 담아 내보냅니다.
    return png_path
                                       # 워드클라우드 이후 이미지 파일로 저징해주는 함수
def save_as_png(wordcloud, png_path):  # wordcloud: txt 파일을 워드클라우드로 변환한 변수, png_path: 이미지 파일로 저장될 경로를 담고있는 변수
    wordcloud.to_file(png_path)        # 워드클라우드된 내용을 png_path 변수의 경로로 저장합니다. 
    print(f'파일이 {png_path} 경로로 저장되었습니다.')
    
def quick_view(wordcloud):             # 저장 단계를 거치지 않고 미리보기를 보여주는 함수 
    plt.figure(figsize = (20,20), dpi=300)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')                    # 의미 없는 그래프 축 제거 
    plt.show()
    
def menu_1():    # 소메뉴1로 사용자에게 워드클라우드 시킨 자료를 바로 저장시킬지 저장시키기전에 미리보여줄지 묻습니다.
    while(True):
        print('파일을 저장하거나 이미지 파일을 바로 확인합니다. 메뉴를 선택해주세요.')
        try:
            menu = int(input('1. 파일 저장 2. 바로 보기'))
            if menu == 1:    # 1. 파일 저장
                return True
            if menu == 2:    # 2. 바로 보기 
                return False 
        except ValueError:   # 1, 2 메뉴 이외의 선택 
            print('존재하지 않는 메뉴입니다.')
            continue
    
def menu_2():   # 소메뉴2로 미리보기 함수 이후에 노출되는 함수입니다. 파일 저장 여부를 묻습니다. 
    while(True):
        save_menu = input('파일을 저장할까요? (Y/N)') 
        if save_menu in ['Y', 'y']:
            print('파일을 저장하겠습니다.')
            return True
        elif save_menu in ['N', 'n']:
            print('파일을 저장하지 않습니다.')
            return False 
        else:
            print('Y 또는 N으로만 입력해주세요.')
            continue
    
def menu_3():   # 소메뉴3으로 파일 저장 이후에 보여지는 함수입니다. 상위 메뉴로 올라가 계속 진행할지 종료할지 묻습니다.
    while(True):
        print('계속 진행하거나 시스템을 종료합니다. 메뉴를 선택해주세요.')
        where_to = input('1. 상위 메뉴 2. 종료')
        if where_to == '1':   # 상위 메뉴로 이동합니다.
            print('상위메뉴를 선택하였습니다.')
            return True 
        elif where_to == '2': # 프로그램을 종료시킵니다.
            print('종료를 선택하였습니다.')
            sys.exit('프로그램을 종료합니다.')
        elif where_to == '': # 아무것도 입력하지 않은 경우입니다. 
            print('존재하지 않는 메뉴입니다.')
            continue

def run_sys():                  # 앞서 정의한 함수들을 순서에 맞춰 워드클라우드 코드를 돌려줍니다. 
    words = open_txt_file()     # txt 파일을 읽어주는 함수 
    desg_mask = open_img_file() # 워드클라우드에 사용할 마스킹 이미지를 불러오는 함수
    path = set_font_path()      # 워드클라우드에 사용할 폰트를 불러오는 함수 
    wordcloud = wordclouder(path, desg_mask, words) # txt 파일을 워드클라우드시키는 함수 path: 폰트 경로, desg_mask: 마스킹 이미지 경로, words: txt 내용(명사) 
    if menu_1() == True:                     # 사용자가 시각화 이후에 파일 저장을 선택한 경우입니다.
        png_path = file_path_for_png()       # 이미지 파일이 저장될 경로를 입력받는 함수
        save_as_png(wordcloud, png_path)     # 이미지 파일을 저장해주는 함수 wordcloud: txt 파일을 워드클라우드로 변환한 변수, png_path: 이미지 파일로 저장될 경로를 담고있는 변수
        if menu_3() == True:                 # 시각화 파일을 저장 이후에 계속 진행할지 묻는 함수로 상위 메뉴를 선택한 경우입니다.
            run_sys()                        # 코드를 처음부터 다시 호출해줍니다.
    else:                                    # 파일 저장을 선택하지 않고 미리 보기를 선택한 경우입니다.
        quick_view(wordcloud)                # 시각화 이미지를 미리 보여주는 함수
        if menu_2() == True:                 # 미리보기 이후에 파일 저장을 선택한 경우입니다. 
            png_path = file_path_for_png()   # 이미지 파일이 저장될 경로를 입력받는 함수
            save_as_png(wordcloud, png_path) # 이미지 파일을 저장해주는 함수 wordcloud: txt 파일을 워드클라우드로 변환한 변수, png_path: 이미지 파일로 저장될 경로를 담고있는 변수
            if menu_3() == True:             # 시각화 파일을 저장 이후에 계속 진행할지 묻는 함수로 상위 메뉴를 선택한 경우입니다. 
                run_sys()                    # 코드를 처음부터 다시 호출해줍니다.
        else:                                # 미리보기 이후에 파일 저장을 선택하지 않은 경우입니다.
            if menu_3() == True:             # 상위 메뉴로 올라가는것을 선택한 경우입니다.
                run_sys()                    # 코드를 처음부터 다시 호출해줍니다.
                
run_sys()
```
